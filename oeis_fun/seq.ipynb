# データ取り込み

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# strippedファイルの５行目からdataframeに変換
df = pd.read_csv('stripped', skiprows=4, header=None, delim_whitespace=True)
df.head()
```

```python
# カラム1の各行をコンマで分割してリストに変換し、特定の整数が含まれるかチェックしてカウントする関数を定義
def count_integer_occurrences(row):
    numbers = [int(num) for num in row.split(',') if num.strip()]  # 空の文字列を除外
    return numbers

# 各行に対して関数を適用して特定の整数の出現回数をカウントし、それを新しい列としてデータフレームに追加
df[2] = df[1].apply(count_integer_occurrences)
# dfの最初の40行を表示
df.head(40)
```

```python
# カラム2全体でどのような整数が何回出現したかをカウントする関数を定義
def count_integer_occurrences(column):
    # 全ての整数を1つのリストにまとめる
    all_integers = [integer for row in column for integer in row]
    # 整数の出現回数をカウントする
    integer_counts = Counter(all_integers)
    return integer_counts

# 整数の出現回数をカウントし、多い順に表示
integer_counts = count_integer_occurrences(df[2])
cnt = 0
for integer, count in integer_counts.most_common():
    if cnt == 20:
        break
    print(f'{integer}: {count}回')
    cnt += 1
```

```python
# 1回しか出現していない整数を抽出し、sorted()関数を使って整数を昇順に並べ替える
unique_integers = sorted([integer for integer, count in integer_counts.items() if count == 1])
# そこそこ0に近い整数を表示
for integer in unique_integers:
    if integer <= 20000 and integer >= -1500:
        print(integer)
```

```python
# 整数がそれぞれ何回出現したかをカウントする関数を定義
def count_integer_occurrences(column, start, end):
    # 全ての整数を1つのリストにまとめる
    all_integers = [integer for row in column for integer in row]
    # 整数の出現回数をカウントする
    integer_counts = Counter(all_integers)
    # 1から100000までの整数がそれぞれ何回出現したかをカウントする
    all_counts = {integer: integer_counts[integer] for integer in range(start, end + 1)}
    return all_counts

# 1000から50000までの整数がそれぞれ何回出現したかをカウントし、結果を描画
integer_occurrences = count_integer_occurrences(df[2], 1000, 10000)
plt.plot(integer_occurrences.keys(), integer_occurrences.values())
plt.xlabel('integer')
plt.ylabel('appearance count')
plt.title('appearance count of integers from 1000 to 50000 in column 1')
plt.show()
```
